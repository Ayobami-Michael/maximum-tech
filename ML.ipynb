```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path="C:/Users/MRS OLAYINKA/Desktop/DATASETS/auto_clean.csv"
```

```python
df=pd.read_csv(file_path)
```

```python
df
```

```python
#using some variables of the dataset(horsepower,engine-size,curb-weight and highway-mpg) to predict a target variable(price)
#i.e Building a Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
x=df[["horsepower","highway-mpg","engine-size","curb-weight"]]
y=df["price"]
```

```python
lm.fit(x,y)
```

```python
yhat=lm.predict(x)
yhat
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
#using the model to predict the "price" of a car with "horsepower"=120,"highway-mpg"=27,"curb-weight"=2600 and "engine-size"=140
```

```python
yhat_price=-15806.624626329205+(53.49574423)*120+(36.05748882)*27+(81.53026382)*140+(4.70770099)*2600
```

```python
yhat_price
```

```python
#approx. $15,241
```

```python
#some model evaluation techniques to determine if the model is a good fit
```

```python
ax1=sns.distplot(df["price"],hist=False,color="r",label="Actual value")
sns.distplot(yhat,hist=False,color="b",label="Fitted values",ax=ax1)
plt.ylim(0,)
```

```python
#the distribution plot shows how well the predicted values(blue) fit the actual values(red)
```

```python
#using numerical measures to determine fit
#(a): MSE
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df["price"],yhat)
```

```python
#(b) R-squared
```

```python
lm.score(x,y)
```

```python
# the value above shows that the model fits the the data by almost 81%
```
